It is possible to step through commands in an AMPL script one command at a time. Single-step
mode is enabled by

```ampl
option single_step n ;
```

where n is a positive integer; it specifies that if the insert level is at most n, AMPL should behave as
though commands -; were inserted before each command: it should read commands from the
standard input until end or other end of file signal (control-D on Unix, control-Z on Windows).
Some special commands may appear in this mode:

```ampl
step n opt   execute the next command, or n commands
skip n opt   skip the next command, or n commands
next n opt   if the next command is an if-then-else or loopingcommand, execute the entire
                       compound command, or n commands, before stopping again
                       (unless the compound command itself specifies commands -;)
cont         execute until the end of all currently nested compoundcommands
                       at the current insert level
```
