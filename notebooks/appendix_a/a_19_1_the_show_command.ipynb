The command

```ampl
show namelist opt redirection opt ;
```

lists all model entities if namelist is not present. It shows each name's declaration if it has one, or
else lists model entities of the kind indicated by the first letters of each name:

```ampl
ch...   ==>   checks           c...    ==> constraints
e...    ==>   environments     f...    ==> functions
o...    ==>   objectives
pr...   ==>   problems         p...    ==> parameters
su...   ==>   suffixes         s...    ==> sets
t...    ==>   tables           v...    ==> variables
```
