Infinity is a predefined parameter; it is the threshold above which upper bounds are considered
absent (i.e., infinite), and -Infinity is the threshold below which lower bounds are considered
absent. Thus given

```ampl
set A;
param Ub{A} default Infinity;
param Lb{A} default -Infinity;
var V {i in A} >= Lb[i], <= Ub[i];
```

components of V for which no Lb value is given in a data section are unbounded below and components
for which no Ub value is given are unbounded above. One can similarly arrange for
optional lower and upper constraint bounds. On computers with IEEE arithmetic (most modern
systems) Infinity is the IEEE ∞ value.
