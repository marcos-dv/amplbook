The shell command provides a temporary escape to the operating system, if such is permitted,
to run commands.

```ampl
shell 'command - line' redirection opt ;
shell redirection opt ;
```

The first version runs command-line, which is contained in a literal string. In the second version,
AMPL invokes an operating-system shell, and control returns to AMPL when that shell terminates.
Before invoking the shell, AMPL writes a list of current options and their values to the file (if any)
named by option shell_env_file. The name of the shell program is determined by option
SHELL.
