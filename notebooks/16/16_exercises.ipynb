**16-5.** The set covering problem, mentioned at the end of [Section 16.2](./16_2_a_scheduling_model.ipynb), can be stated in a general
way as follows. You are given a set S, and certain subsets T 1 , T 2 , . . . , T n of S; a cost is associated
with each of the subsets. A selection of some of the subsets T `j` is said to `cover` S if every member
of S is also a member of at least one of the selected subsets. For example, if

```ampl
S = {1,2,3,4}
```

and

```ampl
T1 = {1,2,4}   T2 = {2,3}   T3 = {1}    T4 = {3,4}    T5 = {1,3}
```

the selections (T1,T2) and (T2,T4,T5) `cover` S, but the selection (T3,T4,T5) does not.
The goal of the set covering problem is to find the least costly selection of subsets that covers S.
Formulate a columnwise linear program that solves this problem for any given set and subsets.
